```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'http://www.analytictech.com/mb021/mlk.htm'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
speech = soup.find_all('p')
```

```python
speech_combined = [p.text for p in speech]
```

```python
string_speech = ' '.join(speech_combined)
```

```python
string_speech[:100]
```

```python
string_speech_cleaned = string_speech.replace('\r\n',' ')
```

```python
import re
```

```python
string_speech_no_punt = re.sub(r'[^\w\s]','',string_speech_cleaned)
```

```python
speech_lower = string_speech_no_punt.lower()
```

```python
re.split(r'[\s+]',speech_lower)
```

```python
speech_broken_out = re.split(r'[\s+]',speech_lower)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(speech_broken_out).value_counts()
```

```python
df.to_csv(r'C:\Users\Giray\Downloads\MLKJ_Speech_counts.csv',header=['Counts'],index_label=['Word'])
```
